## Problem 2: Estimating the Oort constants (A,B)

###### a) Derivation of Oort constants based on rotation curve

The Oort coordinates, A and B, are defined as follows:

$ A \equiv - \frac{1}{2} \left(R \frac{d\Omega}{dR}\right)_{R_0};\,\,\, B \equiv -\left(\Omega + \frac{1}{2}R\frac{d\Omega_z}{dR}\right)_{R_0}$

assuming circular motion, we can relate those contants with the local circular velocity by (B&M Ch.10.3)

$ A = \frac{1}{2} \left( \frac{v_c}{R}-\frac{dv_c}{dR} \right)_{R_0};\,\,\, B = -\frac{1}{2} \left( \frac{v_c}{R}+\frac{dv_c}{dR} \right)_{R_0}$

The underlying assumption (circular motion) will probably be more likely for gas than for stars. The reason is that orbit circularization requires exchange in momentum, which for gas occurs through dissipation, but for stars it is more difficult given the small intensity of tidal effects. If stars formed from gas already in a circular orbit, then there would be no issues, but some stars may form before the disk is fully settled in circular motion. While gas clouds can have large peculiar velocities due to momentum exchange with SN bubbles to which they are more sensitive than stars (see for instance Reid et al. 2014 ApJ 783:130), I would expect that to be the case only for a few outliers.

As for the interpretation (B&M Chp.10.3), A measures the shear in the disk (differential rotation). Effectively, if $\Omega$ is constant (solid disk), A vanishes. B measures the vorticity of the velocity field.

###### b) Derivation of relationship between Oort constants, proper motion in galactic longitude, distance and galactic longitude

Following Binney & Merrifield (Chapters 9 & 10), in which using some geometry (assuming the moving star is in the galactic plane together with the Sun, than the star is much closer to the Sun than the GC...) and the relationship between the rotation curve and the Oort constants, after projecting we arrive to a simple linear relationship

$ \mu_l = \frac{v_t}{d} = B + A cos(2l)$    (1)

This relationship is the one we will use to infer the values of the Oort constants from the Hipparcos data (which will have the same units as $\mu_l$:

$ A = 14.82 \pm 0.84 km \, s^{-1}\, kpc^{-1}; \,\,\, B = -12.37 \pm 0.64 km \, s^{-1}\, kpc^{-1}$

$ A = 3.13 \pm 0.18 mas \, yr^{-1}\, ; \,\,\, B = -2.61 \pm 0.14 mas \, yr^{-1}$

This values seem to come from Feast & Whitelock (1997). Their methodology is based on the following:

i) They use Hipparcos data for a subset of 220 Galactic Cepheid stars identified within the Hip. catalogue

ii) They correct them for reddening according to their period

We see that since the methodology used there is different from the one we are going to use here (using kinematics information of all available stars), we could find systematic differences in the results

###### Fitting a model: leasts square method

The firs methodology we will use is that of the least square fit. We apply it to all 6 subsamples, and compare the results

```python
# We import the modules needed for optimization
import numpy
import scipy.optimize as optimize
# We define the model we want to fit
def model(pars,x):
    b,a = pars
    return b+a*cos(2*x)
# We define the function we want to minimize, in this case, the residual
def res(pars,xdata,ydata, sigma_y):
    return (ydata - model(pars,xdata))/sigma_y
```

```python
# For each subsample, we initialize the parameters at zero, and use the x, y and sigma_y data that apply
# We need to add an additional isotropic velocity dispersion
sigma_v = 25.0    # km/s
sigma_v = 0.210949526672 * sigma_v    # To use mas/yr units (need to multiply by parallax in mas)
# Full sample
pars0 = numpy.array([0.00,0.00])
x = full_sample[:,5]
y = full_sample[:,6]
sigma_y = (full_sample[:,7]**2 + (sigma_v * full_sample[:,4])**2)**0.5
least_full = optimize.leastsq(res, x0=pars0, args=(x,y,sigma_y), full_output=True)
# Full cleaned sample
pars0 = numpy.array([0.00,0.00])
x = full_clean_sample[:,5]
y = full_clean_sample[:,6]
sigma_y = (full_clean_sample[:,7]**2 + (sigma_v * full_clean_sample[:,4])**2)**0.5
least_full_clean = optimize.leastsq(res, x0=pars0, args=(x,y,sigma_y), full_output=True)
# Full MS sample
pars0 = numpy.array([0.00,0.00])
x = ms_full_sample[:,5]
y = ms_full_sample[:,6]
sigma_y = (ms_full_sample[:,7]**2 + (sigma_v * ms_full_sample[:,4])**2)**0.5
least_ms_full = optimize.leastsq(res, x0=pars0, args=(x,y,sigma_y), full_output=True)
# MS cleaned sample
pars0 = numpy.array([0.00,0.00])
x = ms_clean_sample[:,5]
y = ms_clean_sample[:,6]
sigma_y = (ms_clean_sample[:,7]**2 + (sigma_v * ms_clean_sample[:,4])**2)**0.5
least_ms_clean = optimize.leastsq(res, x0=pars0, args=(x,y,sigma_y), full_output=True)
# Disk full sample
pars0 = numpy.array([0.00,0.00])
x = disk_full_sample[:,5]
y = disk_full_sample[:,6]
sigma_y = (disk_full_sample[:,7]**2 + (sigma_v * disk_full_sample[:,4])**2)**0.5
least_disk_full = optimize.leastsq(res, x0=pars0, args=(x,y,sigma_y), full_output=True)
# Disk cleaned sample
pars0 = numpy.array([0.00,0.00])
x = disk_clean_sample[:,5]
y = disk_clean_sample[:,6]
sigma_y = (disk_clean_sample[:,7]**2 + (sigma_v * disk_clean_sample[:,4])**2)**0.5
least_disk_clean = optimize.leastsq(res, x0=pars0, args=(x,y,sigma_y), full_output=True)
# MS disk full sample
pars0 = numpy.array([0.00,0.00])
x = ms_disk_full_sample[:,5]
y = ms_disk_full_sample[:,6]
sigma_y = (ms_disk_full_sample[:,7]**2 + (sigma_v * ms_disk_full_sample[:,4])**2)**0.5
least_ms_disk_full = optimize.leastsq(res, x0=pars0, args=(x,y,sigma_y), full_output=True)
# MS disk cleaned sample
pars0 = numpy.array([0.00,0.00])
x = ms_disk_clean_sample[:,5]
y = ms_disk_clean_sample[:,6]
sigma_y = (ms_disk_clean_sample[:,7]**2 + (sigma_v * ms_disk_clean_sample[:,4])**2)**0.5
least_ms_disk_clean = optimize.leastsq(res, x0=pars0, args=(x,y,sigma_y), full_output=True)
```

###### Fitting a model: bayesian approach

Here we are going to try to estimate the parameters following a bayesian approach (see Adrian's tutorial). To estimate the posterior distribution for the parameters, we sample from some prior distributions using a MCMC algorithm. Since we do not have particular information on the parameters, we assume uniform priors

```python
# We import the modules we will need to run the code (adapted straight from Adrian's tutorial @ 
# http://nbviewer.ipython.org/github/adrn/teaching/blob/master/galactic-dynamics/notebooks/Fit%20a%20line%20to%20data.ipynb)
import numpy as np
import matplotlib.pyplot as plt
import emcee
import triangle
from scipy.stats import multivariate_normal
```

```python
# We define the model we want the data to fit in
def model(pars, x):
    b,a = pars
    return b*cos(2*x) + a
# The probability of the data given the model is the likelihood:
def ln_likelihood(pars, x, y, sigma_y):
    sum_term = 2*np.log(sigma_y) + ((y - model(pars, x))/sigma_y)**2
    return -0.5*(len(x)*np.log(2*np.pi) + np.sum(sum_term))
# We define the priors as uniform
def ln_prior(pars, x, y, sigma_y):
    b,a = pars
    
    p = 0.
    
    b_range = (-15.,15.)
    if b < b_range[0] or b > b_range[1]:
        return -np.inf
    else:
        p += -np.log(b_range[1] - b_range[0])
        
    a_range = (-15.,15.)
    if a < a_range[0] or a > a_range[1]:
        return -np.inf
    else:
        p += -np.log(a_range[1] - a_range[0])
    
    return 0.

def ln_posterior(pars, *args):
    return ln_prior(pars, *args)+ ln_likelihood(pars, *args)
```

```python
# We define the MCMC parameters
nwalkers = 10
nsteps = 2000

p0 = np.zeros((nwalkers,2))
p0[:,0] = np.random.uniform(0,10,nwalkers)
p0[:,1] = np.random.uniform(0,10,nwalkers)

# We are going to use the full sample
x = full_sample[:,5]
y = full_sample[:,6]
sigma_y = (full_sample[:,7]**2 + (sigma_v * full_sample[:,4])**2)**0.5

sampler = emcee.EnsembleSampler(nwalkers, dim=2, lnpostfn=ln_posterior, 
                                args=(x,y,sigma_y), a=1.25)
pos,prob,state = sampler.run_mcmc(p0,nsteps//10)
sampler.reset()
pos,prob,state = sampler.run_mcmc(pos,nsteps)
```
